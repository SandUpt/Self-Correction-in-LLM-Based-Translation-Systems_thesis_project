```python
import pandas as pd
import numpy as np
```

```python
INPUT_FILE = r'C:\Users\SandUpt\Desktop\thesis writing\HPI Server\evaluations\self_correction\qwen_zh_en_70_30_eval2\results.csv'
OUTPUT_FILE = r'C:\Users\SandUpt\Desktop\thesis writing\HPI Server\evaluations\self_correction\qwen_zh_en_70_30_eval2\sampled_100_translations.csv'
RANDOM_SEED = 50
```

```python

df = pd.read_csv(INPUT_FILE)
print(f"Total rows: {len(df)}")
```

```python

sampled_df = df.sample(n=100, random_state=RANDOM_SEED)

sampled_df = sampled_df.reset_index(drop=False)
sampled_df = sampled_df.rename(columns={'index': 'original_row_index'})
```

```python
sampled_df.to_csv(OUTPUT_FILE, index=False)
print(f"Saved {len(sampled_df)} rows to {OUTPUT_FILE}")
```

```python

display(sampled_df[['source', 'reference', 'initial_translation', 'analysis', 'corrected_translation']].head(3))
```
